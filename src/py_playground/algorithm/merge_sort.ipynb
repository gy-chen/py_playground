Demonstrate idea of merge sort.

```
def merge_sort(a):
  """sort using merge sort algorithm
  
  :param a: array like
  :return list:
  """
  if len(a) <= 1:
    return a
  mid = len(a) // 2
  left_half = a[:mid]
  right_half = a[mid:]
  left_half = merge_sort(left_half)
  right_half = merge_sort(right_half)
  return list(_merge(left_half, right_half))

def _merge(left_half, right_half):
  i, j = 0, 0
  l_len, r_len = len(left_half), len(right_half)
  while i < l_len or j < r_len:
    if j >= r_len:
      yield left_half[i]
      i += 1
    elif i >= l_len:
      yield right_half[j]
      j += 1
    elif left_half[i] < right_half[j]:
      yield left_half[i]
      i += 1
    else:
      yield right_half[j]
      j += 1
```

```
merge_sort([5, 4, 3, 2, 1])
```
